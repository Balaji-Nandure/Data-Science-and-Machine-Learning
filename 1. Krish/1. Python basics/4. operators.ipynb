```python
# Arithmetic operators example in Python

a = 10
b = 5

add_result = a + b            # addition
sub_result = a - b            # subtraction
mult_result = a * b           # multiplication
div_result = a / b            # division
floor_div_result = a // b     # floor division
mod_result = a % b            # modulus
power_result = a ** b         # power

print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(mod_result)
print(power_result)
```

```python
# Logical operators example in Python

x = True
y = False

# and operator
result_and = x and y
print(result_and)

# or operator
result_or = x or y
print(result_or)  


# not operator
result_not = not x
print(result_not)
 
print(bool(-1))  # True
print(bool(0))   # False
print(bool(1))   # True
```
