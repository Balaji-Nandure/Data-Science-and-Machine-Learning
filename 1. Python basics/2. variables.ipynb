```python
# variable declaration
a=100 # integer
name="Balaji" # string
is_student=True # boolean

# Naming conventions
# 1. start with a letter or underscore
# 2. can contain letters, numbers and underscores
# 3. case sensitive
# 4. no spaces allowed
# 5. no special characters allowed
# 6. no keywords allowed
```

```python
# python is dynamically typed language
# type of variable is determined at runtime 
```

```python
# typeconversion
a=5
print(type(a))
a=str(a)
print(type(a))
a=float(a)
print(type(a))
a=bool(a)
print(type(a))
a=int(a)
print(type(a))
```

```python
# dynamic typing
var=10
print(type(var)) # int
var="Balaji"
print(type(var)) # str
var=True
print(type(var)) # bool
```
