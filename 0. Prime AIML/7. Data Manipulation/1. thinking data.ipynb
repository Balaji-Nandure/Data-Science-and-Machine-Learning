```python
import json
```

```python
def load_data(file_name):
    with open(file_name, 'r') as file:
        data = json.load(file)
    return data
```

```python
data = load_data("store_data.json")
```

```python
# In data alice name data is duplicate so we need to remove the duplicate data
# inconsistancy in rating data type
# age missing some values

def clean_data(data):
    text_to_number = {"five": 5, "four": 4, "three": 3, "two": 2, "one": 1}
    cleaned_data = []
    unique_users = set()
    for user in data:
        # clean the rating part
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_number:
            user["rating"] = text_to_number[raw_rating]

        # replave the missing data
        # raw_age = user["age"]
        raw_age = user.get("age") # get function return none if the key is not present
        if(raw_age == None):
            user["age"] = None

        # deduplication
        if(user["name"] in unique_users):
            continue
        unique_users.add(user["name"])

        cleaned_data.append(user)
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
# Get meaningful insights from data
# 1. what is average rating
# 2. what % of user give <3 star rating

def get_insights(data):
    # avg rating
    total_rating = 0
    for user in data:
        print(user["rating"])
        total_rating += float(user["rating"])
    avg_rating = total_rating / len(data)
    print(f"Average rating: {avg_rating}")

    # % of user give <3 star rating
    poor_ratings = 0;
    for user in data:
        if(float(user["rating"]) < 3):
            poor_ratings += 1
    percentage = (poor_ratings / len(data)) * 100
    print(f"% of user give <3 star rating: {percentage}%")

    
```

```python
get_insights(data)
```
