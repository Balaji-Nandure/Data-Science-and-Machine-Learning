```python
# Pandas is a data manipulation and analysis library.
# used for data cleaning, data transformation, data visualization, data analysis, and data modeling.

import pandas as pd
```

```python
# series -> it is similar to column in excel
data = pd.Series([51,32,43,74,326])
print(data)

# create series from dictionary (dont forget S is capital)
data = pd.Series({'a':10,'b':20,'c':30})
print(data) #key will be the index

# series of arr with myindexes
data = [10,20,30]
myIndex = ['g', 's', 'b']
print(pd.Series(data, myIndex))
```

```python
# series has singel column.
# Dataframe has multiple columns.
# Dataframe is a table of data with rows and columns.

data = { #4 rows and 3 columns
    'name': ['balaj', 'kumar', 'sai', 'kiran'],
    'age': [20, 21, 22, 23],
    'city': ['new york', 'los angeles', 'chicago', 'san francisco']
}

myindex = ['a', 'b', 'c', 'd']

df = pd.DataFrame(data, myindex) # passing index is optional; 
print(df)

# converting this dataframe to a arr
import numpy as np
print(np.array(df))
```

```python
# Data manipulation with dataframe
data = { #4 rows and 3 columns
    'name': ['balaj', 'kumar', 'sai', 'kiran'],
    'age': [20, 21, 22, 23],
    'city': ['new york', 'los angeles', 'chicago', 'san francisco']
}

df = pd.DataFrame(data)
print(df)

# adding a new column
df['gender'] = ['male', 'female', 'male', 'female']
print(df)

# removing a column
df.drop(columns=['gender'], inplace=True) # this is now a permanent operation because inplace=True
print(df)

# modifying a column
df['age'] = df['age'] + 2
print(df)

```
