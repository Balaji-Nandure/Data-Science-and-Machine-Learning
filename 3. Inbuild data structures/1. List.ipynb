```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits[0])
print(fruits[-1])
print(fruits[1:3]) # slicing -> [start:end:step]
print(fruits[::2]) # slicing with step
print(fruits[::-1]) # slicing with negative step
print(fruits.index("banana")) # finding the index of an element
fruits[2] = "grape" # updating the list
fruits.append("fig") # adding an element to the list
fruits.remove("banana") # removing an element from the list
fruits.pop() # removing the last element from the list
fruits.pop(1) # removing the element at index 1
fruits.clear() # clearing the list
fruits.sort() # sorting the list
fruits.reverse() # reversing the list
print(fruits.count("apple")) # counting the number of elements in the list
fruits.insert(1, "banana") # inserting an element at index 1
```
