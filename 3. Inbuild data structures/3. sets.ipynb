```python
#  set is a collection of items that are unordered and unique.
# set operations
my_set = {1, 2, 3, 4, 5}
my_set2 = {6, 7, 8, 9, 10}
# union
print(my_set | my_set2)
# intersection
print(my_set & my_set2)
# difference
print(my_set - my_set2)
# symmetric difference
print(my_set ^ my_set2)
# subset
print(my_set <= my_set2)
# superset
print(my_set >= my_set2)
# add element
my_set.add(6)
print(my_set)
# remove element
my_set.remove(6)
print(my_set)
# remove element
my_set.remove(6) # if element is not present, it will raise an error
# discard element
my_set.discard(6) # if element is not present, it will not raise an error
# clear set
my_set.clear()

```
