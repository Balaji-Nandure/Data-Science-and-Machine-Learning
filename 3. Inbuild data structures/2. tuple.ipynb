```python
# Tuple is a collection of items that are ordered and immutable.

# Tuple with different data types
mytuple = ("apple", "apple", 1, 2.5, True)
print(mytuple)
print(mytuple[0])
print(mytuple[-1])
print(mytuple[1:3])
print(mytuple[::2])
print(mytuple[::-1])
print(mytuple.index("apple"))
print(mytuple.count("apple"))
```
