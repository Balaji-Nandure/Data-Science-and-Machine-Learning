```python
# map function is used to apply a function to each item in an iterable and return a new list with the results.

# syntax
# map(function, iterable)

def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
print(map(square, numbers)) # this returns a map object
print(list(map(square, numbers))) # this returns a list of the results

# map with multiple iterables
def add(a, b):
    return a + b

numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6, 5] #this 5 will be ignored
print(list(map(add, numbers1, numbers2)))
```

```python
#  apply inbuilt functions using map
print(list(map(len, ["apple", "banana", "cherry"])))

# upper using map
# def upper(s):
#     return s.upper()

print(list(map(str.upper, ["apple", "banana", "cherry"])))
```
