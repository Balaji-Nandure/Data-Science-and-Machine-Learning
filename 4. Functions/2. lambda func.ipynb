```python
# lambda function is a small anonymous function.
# it can have any number of arguments, but can only have one expression.
# lambda function is used when we need a small function to do a simple task.

# syntax
# lambda arguments : expression
# addition lambda function
add = lambda a, b : a + b

print(add(10, 20))
```

```python
#  lambda map function
numbers = [1, 2, 3, 4, 5]
print(list(map(lambda x: x * 2, numbers)))
```
