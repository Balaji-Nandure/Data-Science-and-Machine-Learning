```python
# Data serialization and deserialization
import json
data = {'name': 'Balaji', 'age': 20, 'city': 'Chennai'}

# convert python object to json object
json_data = json.dumps(data)
print("json data", json_data)
print("type of json data", type(json_data))

# convert json object to python object
python_data = json.loads(json_data)
print("python data", python_data)
print("type of python data", type(python_data))
```

```python
# CSV
import csv

with open('data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Balaji', 20, 'Chennai'])
    writer.writerow(['I Love Dipti'])

with open('data.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print("read -> ", row)
```

```python
# regular expression
import re

# earch for one or more digits
pattern = r'\d+'

# Text to search in
text = 'There are 123 apples'

# Search the text for the pattern
match = re.search(pattern, text)

# Print the found group if match is found
if match:
    print(match.group())
```
